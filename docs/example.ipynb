# Example usage

To use `tjspy` in a project:

```python
import tjspy.cjpg
import pandas as pd
import shutil
```

```python
temp_dir = 'cjpg_download'
downloaded_files = tjspy.cjpg.download('lol', dir=temp_dir, pagina_ini=1, pagina_fim=3)

downloaded_files
```

```python
parsed = [tjspy.cjpg.parse([f]) for f in downloaded_files[1:]]
```

```python
dados = pd.concat(parsed)

dados
```

```python
try:
  shutil.rmtree(temp_dir)
except OSError as e:
  print(f"Error: {e}")
```
